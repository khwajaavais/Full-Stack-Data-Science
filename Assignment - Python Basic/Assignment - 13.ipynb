<h3>Q6</h3>
<b>What function takes a string of JSON data and returns a Python data structure?

<pre>Same as json.loads() method :: <b>decode(o)</b> return Python data structure of JSON string or data

```python
import json
# import JSONDecoder class from json

from json.decoder import JSONDecoder
colour_string = '{ "colour": ["red", "yellow"]}'

# directly called decode method of JSON
json_dict = JSONDecoder().decode(colour_string)

display(type(json_dict))
display(json_dict)

json_dict['colour']
```
